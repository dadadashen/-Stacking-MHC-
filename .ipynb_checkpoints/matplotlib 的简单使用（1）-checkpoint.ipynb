### 简单的折线图

```python
import matplotlib.pyplot as plt
squares=[1,4,9,16,25]
plt.plot(squares)
plt.show()
```

```python
import matplotlib.pyplot as plt
input_values=[1,2,3,4,5]
squares=[1,4,9,16,25]
plt.plot(input_values,squares,linewidth=5)#plt.plot(x,y,大小)

#设置图表标题，并给坐标轴加上标签
plt.title("titile",fontsize=24)
plt.xlabel("value",fontsize=24)
plt.ylabel("square of value",fontsize=14)
plt.tick_params(axis='both',labelsize=14)

plt.show()
```

### 用scatter（）绘制散点图

```python
input_values=[1,2,3,4,5]
squares=[1,4,9,16,25]
plt.scatter(input_values,squares,s=5)#plt.plot(x,y,大小)

plt.title("titile",fontsize=24)
plt.xlabel("value",fontsize=24)
plt.ylabel("square of value",fontsize=14)
plt.tick_params(axis='both',labelsize=14)

plt.show()
```

```python
#输入数据部分
x_value=list(range(1,101))
y_value=[x**2 for x in x_value]

#绘制基本的散点图，但是还没有展示出来
plt.scatter(x_value,y_value,edgecolor='none',c=y_value,cmap=plt.cm.Blues,s=40)#plt.plot(x,y,消除轮廓，颜色，颜色的变化，点的大小)
#参数c的值可以是'red'之类，也可以是数值

#设置标题、x轴y轴标签
plt.title("titile",fontsize=24)
plt.xlabel("value",fontsize=24)
plt.ylabel("square of value",fontsize=14)

#设置刻度标准：大小和单位
plt.tick_params(axis='both',labelsize=14)
plt.axis([0,110,0,11000])

# #show
# plt.show()

#保存
plt.savefig('散点图示例.png',bbox_inches='tight')
```
