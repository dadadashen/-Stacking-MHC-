```python
def modify_test(file_name):
#该函数能够将 文本的的制表符通通改成逗号
    #先复制，再清空原本的，最后将修改过的复制内容写入
    with open(file_name, "r+") as file_object:
        read_data=file_object.read()# 用一个临时变量 read_data 存储文本内容
        file_object.seek(0)#把文件定位到position 0，没有这句的话，文件是定位到数据最后，truncate也是从这里删除，所以感觉就是没起作用。
        file_object.truncate()   #清空文件
        file_object.write(read_data.replace('\t', ','))
    
dateset_path='class_II_similarity_reduced_5cv_sep'#数据集的路径
dirs = os.listdir(dateset_path)# os.listdir() 方法用于返回指定的文件夹包含的文件或文件夹的名字的列表。

for file in dirs:    
    '''
        ①删除宿主为mouse的数据。
    '''
    if 'H-2-' in file:
        os.remove(dateset_path+'\\'+file)
        continue

    '''
        ②将文本的的制表符通通改成逗号，目的是为了生成合格的csv文件。将每个文件的后缀从'.txt'更改为‘.csv'
    '''
    modify_test(dateset_path+'\\'+file)
    portion = os.path.splitext(dateset_path+'\\'+file)  # 分离文件名与扩展名
    # 如果后缀是‘.txt'
    if portion[1] == '.txt':
       # 重新组合文件名和后缀名
        newname = portion[0] + '.csv'
        os.rename(dateset_path+'\\'+file, newname)
        
'''
    ③删除不必要的列
'''
dirs = os.listdir(dateset_path)#后缀都改了，所以文件名列表都要更新
for file in dirs:
    # 加载数据集,添加头一行的信息。
    dataframe = pd.read_csv(dateset_path+'\\'+file,header=0,names=('host','allele','len','antigen','description','operator','value'))
    # 删除不必要的属性，不会修改源文件
    dataframe =dataframe.drop(['host','allele','len','antigen','operator'], axis=1)
    # 添加标签信息Boole
    true_value=[]#真实值，非 1 即 -1
    threshold=1500#假设阈值为1500
    for i in dataframe['value']:#根据亲和力大小来判断结合力的强弱。强为1，弱为0
        if(i>=threshold):
            true_value.append(0)
        else:
            true_value.append(1)
    dataframe.insert(1,"Boole",true_value)
    dataframe.to_csv(dateset_path+'\\'+file,header=None)      

'''
    ④将train和test类的文件分别移动到两个文件夹中
'''
dateset_path='class_II_similarity_reduced_5cv_sep'#数据集的路径
to_train_path = dateset_path +'\\'+ 'train'
to_test_path = dateset_path +'\\' + 'test'
#  如果不存在train或test文件夹，则创建
if not os.path.isdir(to_train_path):
     os.makedirs(to_train_path)
if not os.path.isdir(to_test_path):
     os.makedirs(to_test_path)

for file in dirs:
     from_path = dateset_path+'\\'+file
     if '_train_' in file:# 这样的缩进才是正确的，有毒
         shutil.move(from_path, to_train_path)
     else:
         shutil.move(from_path, to_test_path) 
'''
    ⑤因为数据集采用的是5折交叉验证，所以需要将文件夹train和文件夹test的文件分为5份，并分到不同的文件夹中
'''
for k in range(2):# k==0时，处理训练集。k==1 时，处理测试集
    if k == 0:
        data_category='train'
    else:
        data_category='test'
    dateset_path='class_II_similarity_reduced_5cv_sep'+'\\'+data_category
    dirs = os.listdir(dateset_path)

    #文件夹名
    to_0_path = dateset_path +'\\'+ data_category+'0'
    to_1_path = dateset_path +'\\'+ data_category+'1'
    to_2_path = dateset_path +'\\'+ data_category+'2'
    to_3_path = dateset_path +'\\'+ data_category+'3'
    to_4_path = dateset_path +'\\'+ data_category+'4'

    #  如果不存在文件夹，则创建
    if not os.path.isdir(to_0_path):
         os.makedirs(to_0_path)
    if not os.path.isdir(to_1_path):
         os.makedirs(to_1_path)
    if not os.path.isdir(to_2_path):
         os.makedirs(to_2_path)
    if not os.path.isdir(to_3_path):
         os.makedirs(to_3_path)
    if not os.path.isdir(to_4_path):
         os.makedirs(to_4_path)

    for file in dirs:# 开始分类文件
        from_path = dateset_path+'\\'+file
        if '_0.csv' in file:
            shutil.move(from_path, to_0_path)
        elif '_1.csv' in file:
            shutil.move(from_path, to_1_path)
        elif '_2.csv' in file:
            shutil.move(from_path, to_2_path)
        elif '_3.csv' in file:
            shutil.move(from_path, to_3_path)
        elif '_4.csv' in file:
            shutil.move(from_path, to_4_path)
        else:
            continue            

'''
    ⑥将每个文件夹的csv文件合并成一个，并删除多余的文件夹。
    因为每个文件夹的文件合并成一个文件后，就没有必要用一个文件夹存储一个文件，所以把合并了的文件提出来之后，要删除空文件夹
'''
for k in range(2):# k==0时，处理训练集。k==1 时，处理测试集
    if k == 0:
        data_category='train'
    else:
        data_category='test'
        
    folder = os.listdir('class_II_similarity_reduced_5cv_sep'+'\\'+data_category)# 获取文件夹名，方便后面删除空文件夹
    for i in range(5):
        dateset_path='class_II_similarity_reduced_5cv_sep'+'\\'+data_category+'\\'+data_category+str(i)
        dirs = os.listdir(dateset_path)
        read_data='num,description,Boole,value\n'# 加上头行信息
        for file in dirs:
            file_name=dateset_path+'\\'+file
            with open(file_name, "r") as file_object:
                read_data+=file_object.read().rstrip()
            read_data += '\n'
            os.remove(file_name)

        com_csv='class_II_similarity_reduced_5cv_sep'+'\\'+data_category+'\\'+data_category+str(i)+'.csv'# 把合成文件放在train的目录下，本来是 train/i 的
        with open(com_csv, "w") as file_object:
            file_object.write(read_data.rstrip())
        
    for i in folder:#删除空文件夹
        os.rmdir('class_II_similarity_reduced_5cv_sep'+'\\'+data_category+'\\'+i)
    
print('end')
```

# 数据的最后清洗----序列数据的向量化
序列数据的向量化分为两步：一是“获得氨基酸残基的词向量”；二是“特征矩阵映射为特征向量”

###  获得氨基酸残基的词向量

        ①先训练所有的csv文件，获得词向量。
        
        模型的输入要形如：
        [['G', 'E', 'P', 'K', 'G', 'A', 'A', 'E', 'S', 'S', 'S', 'K', 'A', 'A', 'L'],
         ['K', 'L', 'L', 'P', 'V', 'P', 'P', 'T', 'V', 'T', 'I', 'F', 'K', 'I', 'S'],
         ['P', 'F', 'P', 'Q', 'P', 'Q', 'Q', 'P', 'F', 'C', 'Q', 'Q', 'P', 'Q', 'R']]
         每个小列表是某个氨基酸序列的分割串，作为“句子”，而氨基酸残基作为“词”。这是因为word2vec本来就是根据“词”预测上下文或根据上下文预测“词”。所以在进行word2vec训练的时候，“句子”中的“词”应该是相关的。而在一个氨基酸序列中，每个氨基酸残基存在某种关联。但不同的氨基酸序列之间几乎不、也很难证明存在着某种关联。这就像是把“我今天打了一下午篮球”和“生物制药是一门理工科专业”当做两个“词”，这两个“词”构成一个句子。但很明显，这个硬凑出来的句子是毫无意义的。
         我们更需要做的是，告诉woed2vec训练器：“我”“今天”“打了”“一下午”“篮球”这些词语构成了句子，形成某种关联。然后我们希望，这个训练器能够仅仅根据“一下午”预测出上下文“我今天打了___篮球”，又或者仅仅根据“我今天打了___篮球”预测出空位应该是“一下午”。当然，随着投喂的句子越来越多，预测的结果也越来越多：不仅仅是“一下午”，也有可能是“一上午”之类的。 
         一个文件所有的氨基酸序列构成一个大列表，表示一篇“文章”，作为一个语料库。
         ②词向量组成的矩阵是word2vec模型中的一个权重矩阵，故可以从模型中获得词向量，以字典的形式保存。一个“词”对应一个词向量，每个氨基酸残基作为一个“词”，故有20个词向量。由于“词”的个数仅有20个，因此其维度size设置为4是足够的。实验证明，size=4与size=40对实验结果几乎没有影响。

```python
'''
    ①先训练所有的csv文件，获得模型。保存备份模型、删除模型。
'''
# 引入日志配置
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

# 引入数据集
dateset_path='class_II_similarity_reduced_5cv_sep'#数据集的路径
folders = os.listdir(dateset_path)# 获取文件夹
lis=[]
for folder in folders:# train文件夹和test文件夹
    subdateset_path='class_II_similarity_reduced_5cv_sep'+'\\'+folder #数据子集(train和test）的路径
    files = os.listdir(subdateset_path)# 获取文件夹
    for file in files:     
        dataframe=pd.read_csv(subdateset_path+'\\'+file)
        for i in dataframe['description']:
            lis.append(i)# 一个残基一个残基的加入，从而实现切分词汇

# 构建模型
model = word2vec.Word2Vec(lis,size=4,window=5,sg=1,min_count=1)

# 保存模型
model.save('word2vec_s04.model')  

# 删除占用大内存的model
del model 

print('end')
```

```python
'''
    从备份的模型获得词向量，以字典的形式保存。
'''
loaded_model = word2vec.Word2Vec.load('word2vec_s04.model')  # 加载模型
# 将获得的词汇及其对应的向量按字典的格式存放到 word_vector_dict 中
word_vector_dict={}
# print(model.wv.index2word) # 获得所有的词汇
for word in loaded_model.wv.index2word:
    #(word,model[word]) #获得词汇及其对应的向量
    word_vector_dict[word]=list(loaded_model[word])
    
# 展示训练的词及其词向量
# print(len(word_vector_dict))
# print(word_vector_dict)
```

### 特征矩阵映射为特征向量

    特征矩阵来自氨基酸残基的词向量的组合，即：每个氨基酸残基对应一行向量，长度为len的氨基酸序列就对应len*size的矩阵（记为 matrix）。这里size是在训练word2vec模型时设置的，size越大，词向量越精确，但计算量越大、泛化能力越差。
    特征矩阵不能直接用于分类算法的训练，需要转化为特征向量。虽然能够将特征矩阵直接按行展开成特征向量，但是会出现维数过大、维数不唯一（氨基酸序列长度不唯一）的问题。
    在这里使用的是，权重向量*特征矩阵 = 特征向量 的方法。数学表达式为：end_row = weight * matrix = w1*r1+w2*r2+w3*r3+...+（w_len）*r_len 。其中r1,r2,...,r_len 都为行向量，end_row 自然也是行向量，且长度与氨基酸长度无关。
    
    ①将氨基酸序列转换成矩阵。矩阵以列表存储
    ②根据矩阵的行数（由氨基酸序列长度决定），生成对应的权重行向量 weight_row。weight_row是一个等差序列，考虑了序列中残基的前后顺序，而且相近的残基得分不会差太多。也可根据氨基酸残基锚定与非锚定进行调整数值。但目前优先跑完整个流程，后续再改。
    ③权重行向量 weight_row 与 矩阵相乘，得目标特征行向量 end_row。
    ④将（标签值，end_row）这两列数据读入临时文件中，并覆盖同名文件。因为希望最终的数据文件应该只包含（标签值，end_row）这两个有用的信息。
    需要后续做的事有：①优化权重向量，最好是做“在等差的基础上，根据锚定氨基酸重新分配权重”和“在无差别序列的基础上，根据锚定氨基酸重新分配权重”

```python
def weight_row(rows):# 根据氨基酸序列长度生成 等差列表 [0.1,0.2,0.3,...,0.1*rows]。系数值可调整，这里只是预设
    mid=int(rows/2)
    w=[0.1,0.1]#首部2位氨基酸残基认为处于RFP区，末尾2位也是
    beg_len=len(w)#绑定核心的开始
    end_len=rows-beg_len-1#绑定核心的结束
    tem=[]
    if rows%2==0:# 1 1 2 2 1 1 
        for i in range(1,mid+1-beg_len):
            w.append(round(0.3*i,2))
            tem.append(round(0.3*i,2))
        len_tem=len(tem)

        for i in range(1,end_len+1-mid+1):
            w.append(tem[-i])

    else:# 1 1 2 3 2 1 1 
        for i in range(1,mid+1-beg_len+1):
            w.append(round(0.3*i,2))
            if i <mid+1-beg_len:
                tem.append(round(0.3*i,2))
        len_tem=len(tem)
        for i in range(1,end_len+1-mid):
            w.append(tem[-i])

    w+=[0.1,0.1]
    return w

def row_value_change(lt,k):# 行乘以倍数，行以列表表示
    c = []
    for i in range(len(lt)):
        c.append(float(lt[i])*k)
    return c

def list_add(a,b,length):# 实现两个列表对应项相加
    c = []
    for i in range(length):
        c.append(a[i]+b[i])
    return c

''' 
    实现 “ 氨基酸序列 -> 特征矩阵 -> 特征行向量 ” 的映射
'''
n_cols=len(word_vector_dict['L'])# n_cols 的值等于 word2vec.Word2Vec()中的 size 参数。代表的词向量的列数，特征行向量end_row需要用到

def feature_rows(path):# 将给定文件中的氨基酸序列组转换为行向量组，返回行向量组（列表形式）
    aa_sequences = pd.read_csv(path,usecols=[1])# 获取氨基酸序列（位于第二列），aa 即 氨基酸aminoacid
    feature_rows=[] #用于收集
    for aa in aa_sequences['description']: # 遍历每个氨基酸序列
    #     print(aa)
        n_rows=len(aa)#15
        # 由序列和词向量生成对应的矩阵。矩阵以列表存储
        '''
            ①将氨基酸序列转换成矩阵。矩阵以列表存储。
        '''
        matrix=[]
        for a in aa:# 遍历每个氨基酸残基
    #         print(a,end=' ')
    #         print(word_vector_dict[a])
            matrix.append(word_vector_dict[a])
    #     print("matrix=",end=" ")
    #     print(matrix)

        '''
            ②根据矩阵的行数（由氨基酸序列长度决定），生成对应的权重行向量 weight_row
        '''
        w_row=weight_row(n_rows)# 返回一个（1*len(aa)）的权重行向量
    #     print("w_row=",end=" ")
    #     print(w_row)

        '''
            ③权重行向量 weight_row 与 特征矩阵 matrix 相乘，得目标特征行向量 end_row。
            数学表达式为：end_row = weight * matrix = 0.1*r1+0.2*r2+0.3*r3+...+（0.1*len）*r_len 。
            其中r1,r2,...,r_len 都为行向量，end_row 自然也是行向量。
        '''
        end_row=[0]*n_cols#初始化
        for i in range(n_rows):
            change_value_row=row_value_change(matrix[i],w_row[i])#行向量乘以系数
            end_row=list_add(end_row,change_value_row,n_cols)#行向量累加

    #     print('end_row=',end=' ')
    #     print(end_row)
    #     print('')
        for i in range(len(end_row)):# 降低特征向量中元素的位数，从而降低后续的计算量。但效果不是很明显
            end_row[i]=round(end_row[i],4)# 仅仅保留4位小数
        feature_rows.append(end_row)
    return feature_rows

'''
    ④将（标签值，end_row）这两列数据读入临时文件中，并覆盖同名文件。
    最终的数据文件应该只包含（标签值，end_row）这两个有用的信息。但是使用pandas工具处理csv的时候，始终会有下标。
    所以最终的文件包含（下标，标签值，end_row）
'''

'''
    先弄train数据集
'''
dateset_path='class_II_similarity_reduced_5cv_sep'+'\\'+'train'
dirs = os.listdir(dateset_path)
for file in dirs:
    file_name=dateset_path+'\\'+file
    file = pd.read_csv(file_name,usecols=[2])# 清洗完毕的file文件，只包含（end_row,标签值）
    file.insert(1,"feature_row",feature_rows(file_name))
    file.to_csv(file_name)# header参数 根据需要更改
   
    '''
        再弄test数据集
    '''
dateset_path='class_II_similarity_reduced_5cv_sep'+'\\'+'test'
dirs = os.listdir(dateset_path)
for file in dirs:
    file_name=dateset_path+'\\'+file
    file = pd.read_csv(file_name,usecols=[2])# 清洗完毕的file文件，只包含（end_row,标签值）
    file.insert(1,"feature_row",feature_rows(file_name))
    file.to_csv(file_name)# header参数 根据需要更改
    
print('end')
```

```python
'''
    进行交叉验证训练之前，将分散的训练数据和测试数据合并，方便进行试验。
'''
def merge_csv(com_train,com_test):#将train好test文件夹的csv文件分别合并成一个,保存路径为 com_train,com_test    
    for i in range(2):
        if i == 0:
            dateset_path='class_II_similarity_reduced_5cv_sep'+'\\'+'train'
            save_path=com_train
        else:
            dateset_path='class_II_similarity_reduced_5cv_sep'+'\\'+'test'
            save_path=com_test
            
        dirs = os.listdir(dateset_path)
        read_data=',Boole,feature_row\n'
        for file in dirs:
            file_name=dateset_path+'\\'+file
            with open(file_name, "r") as file_object:
                read_data+=file_object.read().rstrip()
            read_data += '\n'
        read_data=read_data.replace(',Boole,feature_row','')
        read_data=',Boole,feature_row\n'+read_data.lstrip()
        with open(save_path, "w") as file_object:#写入新文件
            file_object.write(read_data.rstrip())
        
merge_csv('tra_s04.csv','tes_s04.csv')#得到合并后的训练集和测试集
```

# 训练模型

```python
def str2list(x):#将 字符串化的列表转换为字符串
    x=eval(x)
    return x

'''
    生成训练数据
'''
train_file=pd.read_csv('tra_s04.csv')
X_train=np.array(list(train_file['feature_row'].apply(str2list)))
y_train=np.array(list(train_file['Boole']))

'''
    生成测试数据
'''
test_file=pd.read_csv('tes_s04.csv')
X_test=np.array(list(test_file['feature_row'].apply(str2list)))
y_test=np.array(list(test_file['Boole']))


def show_auc(model,test_data,test_label,title_name): #画出 ROC 曲线（receiver operating characteristic curve）
    pred = model.predict_proba(np.array(test_data))[:,1]
    fpr, tpr, threshold = metrics.roc_curve(test_label, pred)
    roc_auc = metrics.auc(fpr, tpr)
    plt.figure(figsize=(6,6))
    plt.title(title_name)
    plt.plot(fpr, tpr, 'b', label = 'Val AUC = %0.3f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    print('AUC= ',roc_auc)
```

### 开始训练

```python

### 5折stacking
n_folds = 5
skf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=1)#
aucs=[]
for i,clf in enumerate(clfs):
    print("分类器：{}".format(clf))
    X_stack_test_n = np.zeros((X_test.shape[0], n_folds))#
    print(X_stack_test_n.shape)#10*(1*5) ,10是训练集函数，5是交叉验证数量
    tem_auc=0
    for j,(train_index,test_index) in enumerate(skf.split(X_train,y_train)):
        tr_x = X_train[train_index]
        tr_y = y_train[train_index]
        clf.fit(tr_x, tr_y)
        
        #每得到一个模型就可以进行一次对测试集的预测
        pred = clf.predict_proba(X_test)[:,1]
        tem_auc+=roc_auc_score(y_test,pred)
#         show_auc(model,test_data,test_label,title_name)
        
        #生成stacking训练数据集
        X_train_stack [test_index, i] = clf.predict_proba(X_train[test_index])[:,1]#对验证集的预测
        X_stack_test_n[:,j] = clf.predict_proba(X_test)[:,1]
        
    
    aucs.append(tem_auc/5)
    print(tem_auc/5)
    
    #生成stacking测试数据集
    X_test_stack[:,i] = X_stack_test_n.mean(axis=1) 
    
#按序输出各分类器的auc
aucs
```
