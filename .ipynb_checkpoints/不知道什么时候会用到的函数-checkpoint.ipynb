```python
def find_TorF(path):# 获得某一文件的正类和负类分组
    file = pd.read_csv(path)
    tl=[]# true label
    fl=[]# false label
    def true_false_sep(series):
        la = series["label"]
        fe = series["feature"]
        if la==1:
            tl.append(eval(fe))
        else:
            fl.append(eval(fe))     
    file.apply(true_false_sep,axis=1)#实现分组
    return tl,fl
    
trues,falses=TorF('text.csv')
trues_label=[1]*len(trues)
falses_label=[0]*len(falses)
print(trues)
print(trues_label)
print(falses)
print(falses_label)
```
