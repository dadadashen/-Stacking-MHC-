# 数据集的初步处理

### 需要导入的模块或函数

```python
import os, sys
from os import listdir, getcwd
from os.path import join
import shutil
import pandas as pd
import datetime
from gensim.models import word2vec 

from sklearn import svm
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

from sklearn.datasets import load_breast_cancer
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
import pylab as plt
import warnings;warnings.filterwarnings('ignore')

from  sklearn.datasets  import  make_hastie_10_2
from xgboost.sklearn import XGBClassifier


def weight_row(rows):# 根据氨基酸序列长度生成 等差列表 [0.1,0.2,0.3,...,0.1*rows]。系数值可调整，这里只是预设
    w=[]
    for i in range(1,rows+1):
        w.append(round(0.3*i,2))
    return w

def row_value_change(lt,k):# 行乘以倍数，行以列表表示
    c = []
    for i in range(len(lt)):
        c.append(float(lt[i])*k)
    return c

def list_add(a,b,length):# 实现两个列表对应项相加
    c = []
    for i in range(length):
        c.append(a[i]+b[i])
    return c

''' 
    实现 “ 氨基酸序列 -> 特征矩阵 -> 特征行向量 ” 的映射
'''
n_cols=len(word_vector_dict['L'])# n_cols 的值等于 word2vec.Word2Vec()中的 size 参数。代表的词向量的列数，特征行向量end_row需要用到

def feature_rows(path):# 将给定文件中的氨基酸序列组转换为行向量组，返回行向量组（列表形式）
    aa_sequences = pd.read_csv(path,usecols=[1])# 获取氨基酸序列（位于第二列），aa 即 氨基酸aminoacid
    feature_rows=[] #用于收集
    for aa in aa_sequences['description']: # 遍历每个氨基酸序列
    #     print(aa)
        n_rows=len(aa)#15
        # 由序列和词向量生成对应的矩阵。矩阵以列表存储
        '''
            ①将氨基酸序列转换成矩阵。矩阵以列表存储。
        '''
        matrix=[]
        for a in aa:# 遍历每个氨基酸残基
    #         print(a,end=' ')
    #         print(word_vector_dict[a])
            matrix.append(word_vector_dict[a])
    #     print("matrix=",end=" ")
    #     print(matrix)

        '''
            ②根据矩阵的行数（由氨基酸序列长度决定），生成对应的权重行向量 weight_row
        '''
        w_row=weight_row(n_rows)# 返回一个（1*len(aa)）的权重行向量
    #     print("w_row=",end=" ")
    #     print(w_row)

        '''
            ③权重行向量 weight_row 与 特征矩阵 matrix 相乘，得目标特征行向量 end_row。
            数学表达式为：end_row = weight * matrix = 0.1*r1+0.2*r2+0.3*r3+...+（0.1*len）*r_len 。
            其中r1,r2,...,r_len 都为行向量，end_row 自然也是行向量。
        '''
        end_row=[0]*n_cols#初始化
        for i in range(n_rows):
            change_value_row=row_value_change(matrix[i],w_row[i])#行向量乘以系数
            end_row=list_add(end_row,change_value_row,n_cols)#行向量累加

    #     print('end_row=',end=' ')
    #     print(end_row)
    #     print('')
        for i in range(len(end_row)):# 降低特征向量中元素的位数，从而降低后续的计算量。但效果不是很明显
            end_row[i]=round(end_row[i],4)# 仅仅保留4位小数
        feature_rows.append(end_row)
    return feature_rows

def merge_csv(com_train,com_test):#将train好test文件夹的csv文件分别合并成一个,保存路径为 com_train,com_test 
    a=b=''
    for i in range(2):
        if i == 0:
            dateset_path='tem_out'+'\\'+'train'
            save_path=com_train
            a=save_path
        else:
            dateset_path='tem_out'+'\\'+'test'
            save_path=com_test
            b=save_path
            
        dirs = os.listdir(dateset_path)
        read_data=',Boole,feature_row\n'
        for file in dirs:
            file_name=dateset_path+'\\'+file
            with open(file_name, "r") as file_object:
                read_data+=file_object.read().rstrip()
            read_data += '\n'
        read_data=read_data.replace(',Boole,feature_row','')
        read_data=',Boole,feature_row\n'+read_data.lstrip()
        with open(save_path, "w") as file_object:#写入新文件
            file_object.write(read_data.rstrip())
    return a,b

def show_auc(model,test_data,test_label,title_name): #画出 ROC 曲线（receiver operating characteristic curve）
    pred = model.predict_proba(np.array(test_data))[:,1]
    fpr, tpr, threshold = metrics.roc_curve(test_label, pred)
    roc_auc = metrics.auc(fpr, tpr)
    plt.figure(figsize=(6,6))
    plt.title(title_name)
    plt.plot(fpr, tpr, 'b', label = 'Val AUC = %0.3f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    print('AUC= ',roc_auc)
    
def str2list(x):#将 字符串化的列表转换为字符串
    x=eval(x)
    return x
```

### 生成数据

```python
sizes=[]
for i in range(101):
    if i ==0:
        continue
    else:
        sizes.append(i)
        
for i in range(100):
    '''
        ①先训练所有的csv文件，获得模型。保存备份模型、删除模型。
    '''
    # 引入数据集
    dateset_path='tem_in'#数据集的路径
    folders = os.listdir(dateset_path)# 获取文件夹
    lis=[]
    for folder in folders:# train文件夹和test文件夹
        subdateset_path='tem_in'+'\\'+folder #数据子集(train和test）的路径
        files = os.listdir(subdateset_path)# 获取文件夹
        for file in files:     
            dataframe=pd.read_csv(subdateset_path+'\\'+file)
            for k in dataframe['description']:
                lis.append(k)# 一个残基一个残基的加入，从而实现切分词汇
    # 构建模型
    s=sizes[int(i)]
    model = word2vec.Word2Vec(lis,size=s,window=5,sg=1,min_count=1)
    word_vector_dict={}
    # 获得所有的词汇
    for word in model.wv.index2word:
        #(word,model[word]) #获得词汇及其对应的向量
        word_vector_dict[word]=list(model[word])

    n_cols=len(word_vector_dict['L'])# n_cols 的值等于 word2vec.Word2Vec()中的 size 参数。代表的词向量的列数，特征行向量end_row需要用到
    dateset_path='tem_in'+'\\'+'train'
    if not os.path.isdir('tem_out'+'\\'+'train'):
        os.makedirs('tem_out'+'\\'+'train')
    dirs = os.listdir(dateset_path)
    for file in dirs:
        file_name=dateset_path+'\\'+file
        file_data = pd.read_csv(file_name,usecols=[2])# 清洗完毕的file文件，只包含（end_row,标签值）
        file_data.insert(1,"feature_row",feature_rows(file_name))
        file_data.to_csv('tem_out'+'\\'+'train'+'\\'+file)# header参数 根据需要更改

        '''
            再弄test数据集
        '''
    dateset_path='tem_in'+'\\'+'test'
    if not os.path.isdir('tem_out'+'\\'+'test'):
        os.makedirs('tem_out'+'\\'+'test')
    dirs = os.listdir(dateset_path)
    for file in dirs:
        file_name=dateset_path+'\\'+file
        file_data = pd.read_csv(file_name,usecols=[2])# 清洗完毕的file文件，只包含（end_row,标签值）
        file_data.insert(1,"feature_row",feature_rows(file_name))
        file_data.to_csv('tem_out'+'\\'+'test'+'\\'+file)# header参数 根据需要更改

    '''
        进行交叉验证训练之前，将分散的训练数据和测试数据合并，方便进行试验。
    '''
    #得到合并后的训练集和测试集的文件名
    train_data_name,test_data_name=merge_csv('dif_size_embedding'+'\\'+'train_s'+str(sizes[i])+'.csv','dif_size_embedding'+'\\'+'test_s'+str(sizes[i])+'.csv')


```

### 训练

```python

auc_out=[]
run_time=[]
for i in range(100):
    
    '''
        进行交叉验证训练之前，将分散的训练数据和测试数据合并，方便进行试验。
    '''
    #得到合并后的训练集和测试集的文件名
    train_data_name,test_data_name='dif_size_embedding'+'\\'+'train_s'+str(sizes[i])+'.csv','dif_size_embedding'+'\\'+'test_s'+str(sizes[i])+'.csv'

    '''a
        生成训练数据
    '''
    train_file=pd.read_csv(train_data_name)
    train_data=list(train_file['feature_row'].apply(str2list))
    train_label=list(train_file['Boole'])
    '''
        生成测试数据
    '''
    test_file=pd.read_csv(test_data_name)
    test_data=list(test_file['feature_row'].apply(str2list))
    test_label=list(test_file['Boole'])
    
    start = datetime.datetime.now()
    
    xgb_clf = XGBClassifier(
    silent=0 ,#设置成1则没有运行信息输出，最好是设置为0.是否在运行升级时打印消息。
    #nthread=4,# cpu 线程数 默认最大
    learning_rate= 0.3, # 如同学习率
    min_child_weight=1, 
    # 这个参数默认是 1，是每个叶子里面 h 的和至少是多少，对正负样本不均衡时的 0-1 分类而言
    #，假设 h 在 0.01 附近，min_child_weight 为 1 意味着叶子节点中最少需要包含 100 个样本。
    #这个参数非常影响结果，控制叶子节点中二阶导的和的最小值，该参数值越小，越容易 overfitting。
    max_depth=6, # 构建树的深度，越大越容易过拟合
    gamma=0,  # 树的叶子节点上作进一步分区所需的最小损失减少,越大越保守，一般0.1、0.2这样子。
    subsample=1, # 随机采样训练样本 训练实例的子采样比
    max_delta_step=0,#最大增量步长，我们允许每个树的权重估计。
    colsample_bytree=1, # 生成树时进行的列采样 
    reg_lambda=1,  # 控制模型复杂度的权重值的L2正则化项参数，参数越大，模型越不容易过拟合。
    #reg_alpha=0, # L1 正则项参数
    #scale_pos_weight=1, #如果取值大于0的话，在类别样本不平衡的情况下有助于快速收敛。平衡正负权重
    #objective= 'multi:softmax', #多分类的问题 指定学习任务和相应的学习目标
    #num_class=10, # 类别数，多分类与 multisoftmax 并用
    n_estimators=100, #树的个数
    seed=1000 #随机种子
    #eval_metric= 'auc'
    )
    xgb_clf.fit(np.array(train_data),np.array(train_label),eval_metric='auc')
    #设置验证集合 verbose=False不打印过程
    # xgb_clf.fit(X_train, y_train,eval_set=[(X_train, y_train), (X_test, y_test)],eval_metric='auc',verbose=False)
#     show_auc(xgb_clf,test_data,test_label,'XGB')
    pred = xgb_clf.predict_proba(np.array(test_data))[:,1]
    fpr, tpr, threshold = metrics.roc_curve(test_label, pred)
    auc_out.append(metrics.auc(fpr, tpr))
    
    end = datetime.datetime.now()
    ii=(end-start).total_seconds()
    run_time.append(ii)
```

### 画图

```python
import matplotlib.pyplot as plt

x_data = sizes
y_data = auc_out

plt.tick_params(axis='both', which='major', labelsize=14)

# 设置图表标题并给坐标轴加上标签
# plt.title("XGB_AUC", fontsize=24) 
plt.xlabel("word_size", fontsize=16) 
plt.ylabel("AUC", fontsize=16)

plt.plot(x_data,y_data)
plt.show()
```
