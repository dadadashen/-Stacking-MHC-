```python
def str2list(x):# 
    x=eval(x)
    return x



def show_auc(model,test_data,test_label,title_name): # ROC receiver operating characteristic curve
    pred = model.predict_proba(np.array(test_data))[:,1]
    fpr, tpr, threshold = metrics.roc_curve(test_label, pred)
    roc_auc = metrics.auc(fpr, tpr)
    plt.figure(figsize=(6,6))
    plt.title(title_name)
    plt.plot(fpr, tpr, 'b', label = 'Val AUC = %0.3f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    print('AUC= ',roc_auc)
```

```python

### 5stacking
n_folds = 5
skf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=1)#
aucs=[]
for i,clf in enumerate(clfs):
    print("{}".format(clf))
    X_stack_test_n = np.zeros((X_test.shape[0], n_folds))#
    print(X_stack_test_n.shape)#10*(1*5) ,105
    tem_auc=0
    for j,(train_index,test_index) in enumerate(skf.split(X_train,y_train)):
        tr_x = X_train[train_index]
        tr_y = y_train[train_index]
        clf.fit(tr_x, tr_y)
        
        pred = clf.predict_proba(X_test)[:,1]
        tem_auc+=roc_auc_score(y_test,pred)
        #stacking
        X_train_stack [test_index, i] = clf.predict_proba(X_train[test_index])[:,1]
        X_stack_test_n[:,j] = clf.predict_proba(X_test)[:,1]
    
    aucs.append(tem_auc/5)
    print(tem_auc/5)
    
    #stacking
    X_test_stack[:,i] = X_stack_test_n.mean(axis=1) 
    
#auc
aucs
```

```python
###LRauc
clf_second = LogisticRegression(solver="lbfgs")
clf_second.fit(X_train_stack,y_train)
pred = clf_second.predict_proba(X_test_stack)[:,1]
roc_auc_score(y_test,pred)#0.9946
```
